# Harjoitustyö - Johdanto datatieteeseen 2021

```python
import pandas as pd

#Haetaan Kööpenhaminen AirBnB-data muuttujaan tiedoston URL:a hyödyntämällä
link = 'http://data.insideairbnb.com/denmark/hovedstaden/copenhagen/2021-02-25/data/listings.csv.gz'

#Jotta csv.gz tiedosto voidaan lukea tulee käyttää compression metodia
data = pd.read_csv(link, compression='gzip')
```

```python
#Jotta pystytään tarkastelemaan sarakkeita laajennetaan näkymä kaikkiin
#sarakkeisiin
pd.set_option('max_columns', None)
data.head()
```

```python
#Otetaan käyttöön vain harjoitustyön kannalta keskeisimmät sarakkeet
data = data[['id', 
             'neighborhood_overview', 
             'host_id',
             'host_about',
             'host_response_time', 
             'neighbourhood_cleansed',
             'room_type',
             'accommodates',
             'bedrooms',
             'beds',
             'amenities',
             'price', 
             'availability_365',
             'number_of_reviews',
             'review_scores_rating',
             'review_scores_accuracy',
             'review_scores_cleanliness',
             'review_scores_checkin',
             'review_scores_communication',
             'review_scores_location',
             'review_scores_value']]
```

```python
data.head()
```

```python
data.info()
```

```python
print(data.isnull().sum())
```

```python
#Poistetaan null-arvot
data = data[data.bedrooms != 0]
data = data[data.beds != 0]
data = data[data.price != 0]
data = data[data.accommodates != 0]
data = data.dropna(axis=0)
```

```python
data.info()
```

```python
#Poistetaan hinnasta $-merkki ja muutetaan se liukuluvuksi
data['price'] = data['price'].replace('[\$,)]','', regex=True).replace('[()]','-', regex=True).astype(float)
```

```python
print(data['price'])
```

```python
#Katsotaan datasetin jakaumaa makuuhuoneiden määrän suhteen
data.bedrooms.hist(bins=50)
```

```python
data.room_type.hist(bins=50)
```

```python
#Keskitytään pelkästään kokonaisiin asuntoihin sekä 1-4 huoneen asuntoihin
data = data[data.room_type != 'Private room']
data = data[data.room_type != 'Hotel room']
data = data[data.bedrooms < 5]
data = data[data.price < 10000.0]
```

```python
#Huomataan, että neighbourhood_cleansed, room_type sekä host_response_time ovat merkkijonoja
#Käytetään get_dummies metodia ja muunnetaan ne dummy-muuttujiksi, jotta niiden käsittely on myöhemmin helpompaa

#Kategorisoidaan majoitusten naapurusto ja huonetyypit
n_dummies = pd.get_dummies(data.neighbourhood_cleansed)
rt_dummies = pd.get_dummies(data.room_type)
hrt_dummies = pd.get_dummies(data.host_response_time)

#Korvataan vanhat sarakkeet uusilla int-arvoilla
alldata = pd.concat((data.drop(['neighbourhood_cleansed', \
    'room_type', 'host_response_time'], axis=1), \
    n_dummies.astype(int), rt_dummies.astype(int), \
    hrt_dummies.astype(int)), \
    axis=1)
allcols = alldata.columns
alldata.head(5)
```

```python
import numpy as np
from sklearn import linear_model
from sklearn.linear_model import LinearRegression

#Lineaarinen regressio kokonaisarvion ja hinnan suhteelle
lr_review = linear_model.LinearRegression()
x = data['review_scores_rating'].values[:, np.newaxis]
y = data['price']

classifier = lr_review.fit(x, y)
plt.scatter(x, y)
plt.plot(x, classifier.predict(x), color = 'red')
plt.xlabel('Kokonaisarvio')
plt.ylabel('Hinta')
plt.show
```

```python
#Lineaarinen regressio saatavuuden ja hinnan suhteelle
lr_availability = linear_model.LinearRegression()
x = data['availability_365'].values[:, np.newaxis]
y = data['price']

classifier = lr_availability.fit(x, y)
plt.scatter(x, y)
plt.plot(x, classifier.predict(x), color = 'red')
plt.xlabel('Saatavuus')
plt.ylabel('Hinta')
plt.show
```

```python
#Lineaarinen regressio makuupaikkojen ja hinnan suhteelle
lr_accommodates = linear_model.LinearRegression()
x = data['accommodates'].values[:, np.newaxis]
y = data['price']

classifier = lr_accommodates.fit(x, y)
plt.scatter(x, y)
plt.plot(x, classifier.predict(x), color = 'red')
plt.xlabel('Makuupaikat')
plt.ylabel('Hinta')
plt.show
```
