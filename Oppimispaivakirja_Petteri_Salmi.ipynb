Data-analytiikan hyödyntämiselle keskeistä on yhdistää se liiketoimintaan sekä tunnistaa sen tarve eri liiketoiminnan osa-alueissa. Luennolla esitetyn analytiikan nelikentän (kuvaileva, diagnosoiva, ennakoiva ja ohjaava analytiikka) avulla on mahdollista jakaa analytiikan tarve tietyille liiketoiminnan osa-alueille. Seuraavassa listauksessa yhdistän kunkin analytiikan teeman liiketoiminnan osa-alueeseen, jotka ovat sopivia analytiikan teemoille.

1. Kuvaileva analytiikka: Tuotannon läpimenoajan kehitys vuoden aikana

3. Ennakoiva analytiikka: Millainen on tuotannon oppimiskäyrä, millä kapasiteetilla voidaan tuottaa tuotetta vuoden päästä
4. Ohjaava analytiikka: Mitä pitäisi tuotannossa kehittää, jotta läpimenoaika pienenee (esim. tietty vaihe)

Bernard (2017) käsittelee blogissaan (https://benbernardblog.com/web-scraping-and-crawling-are-perfectly-legal-right/) datan keräämisen laillisuutta. Hän esittelee tapauksia, jotka ovat johtaneet sanktioihin sekä tapauksia, jotka ovat datan laillisen hyödyntämisen ja laittomuuden välillä. Se herättääkin eettisiä kysymyksiä datan hyödyntämisestä. Laaksonen & Salonen (2018) käsittelevät kirjoituksessaan (https://rajapinta.co/2018/12/04/kuka-saa-paattaa-mita-dataa-tutkijalla-on-kaytossaan-ei-ainakaan-amerikkalainen-suuryritys/) yleisesti aineistojen saatavuutta ja datan keräämistä sekä peilaavat niitä GDPR:n ja tietosuojalain laatimiin rajoitteisiin. Tekstissä nostetaan esiin ristiriita lain ja eettisen oikeuden välillä. Laki tarjoaa mahdollisuuden hyödyntää henkilötietoja tieteellisessä tutkimuksessa, mutta voidaan pohtia onko se eettisesti oikein.

Kumpikin teksti herättää pohdintaa varsinkin datan keräämisen eettisyydestä. Mikäli datalla pystytään vahingoittamaan jonkun liiketoimintaa, aiheuttamaan pahennusta tai pyritään vain taloudelliseen hyötyyn on syytä pohtia omia eettisiä valintoja. Joka tapauksessa datan kerääminen vaatii niin lain tuntemusta kuin maalaisjärjen käyttöä sekä eettistä ymmärrystä.

Luennolla esitetty esimerkki kielen analysoinnista vaikutti mielenkiintoiselta, joten päätin demota sitä seuraavaksi, mutta suomen kielisillä stopwordseillä.

```python
arvio = "Minulla on ollu erinäinen määrä kiinalaisia halpagrillejä, jotka palavat puhki kesässä tai kahdessa ja joissa ei lämpöä oikein riitä. Weber ei ole halpa enkä vielä tiedä kuinka kauan kestää mutta lämpöä ainakin riittää. Kasaaminen oli helppoa kun tarvinut arpoa osien kanssa tai väännellä kohdalleen. Grillaaminenkin sujuu kun tottuu siihen ettei tarvitse pitää nuppeja maksimiasennossa."
arvio
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
tokenArvio=word_tokenize(arvio)
print(tokenArvio)
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopword=set(stopwords.words("finnish"))
print(stopwords)
```

```python
#Lisätään listaan hyödyllisiä sanoja, kun stopwordsit ovat poistettu
putsattu_arvio = []
for sana in tokenArvio:
    if sana not in stopword:
        putsattu_arvio.append(sana)
putsattu_arvio
```

Ohajaamattomassa oppimisessa ei voida tehdä mallia, jolle annetaan syöte ja se tuottaa oikean vastauksen. Mallin toimivuus on myös vain arvailtavissa. Ohjaamattomassa koneoppimisessa pyritään löytämään datasta rakenne. Mallin tavoitteena on löytää säännönmukaisuuksia datasta, jotta tuloksia voidaan luoda. Ohjaamaton koneoppiminen jaetaan erilaisiin menetelmiin. Tässä oppimispäiväkirjassa keskitytään aihemallinnukseen ja ryvästämiseen.

On hyvä kuitenkin pohtia myös ohjaamattoman oppimisen eettistä näkökulmaa. Vaiste (2017) nostaa blogissaan esille ohjaamattoman oppimisen mallin toimivuuuden, sillä oikeaa ratkaisua ei välttämättä ole. Väärä malli johtaa vääriin toimenpiteisiin. Eettinen kysymys herääkin mallin valinnassa, datan vääristymättömyydestä sekä tuloksien tulkinnassa. Linkki materiaaliin: https://juhovaiste.fi/fi/koneoppimisen-muodot-ja-niiden-eettiset-n%C3%A4k%C3%B6kulmat 

Mielestäni tässäkin on syytä korostaa maalaisjärkeä, mikäli sen käyttö on mahdollista. Aihemallinnus arvioi datasta kahta jakaumaa: aiheiden jakaumaa aineistossa ja sanojen jakaumaa aiheissa. Aihemallinnus perustuu oletukselle, että aineisto on aiheiden yhdistelmä ja sanoilla on niille yksilöllinen todennäköisyys kuulua jokaiseen aiheeseen. Yhteenvetona voidaan todeta, että aihemallinnus käsittelee aineiston aiheita ja sanoja todennäköisyysjakauman avulla. Käyttäjän ainoa määriteltävä parametri on käsiteltyjen aiheiden määrä, joka valitaan usein niiden tulkittavuuden tai yhtenäisyyden perusteella. Aihemallinnuksen tuloksena on aiheiden sekä siihen liittyvien merkittävimpien sanojen yhdistelmä, josta voidaan tehdä aineistoon liittyviä päätelmiä sekä jatkotutkimuskysymyksiä. (Toivanen et al. 2020) Linkki materiaaliin: https://doi.org/10.23983/mv.91078

Mielestäni Fryn (2008) seitsemän askelta kuvaa mainiosti välivaiheita kohti laadukasta visualisointia. 

Luennolla esitetyt visualisoinnin yleiset periaatteet ovat hyvä muistaa datatiedeprosessissa. Mielestäni tärkein kolmesta ohjeesta on: tunne yleisösi, sillä sen vaikutus on suurin visualisoinnin tekemiseen sekä siitä aiheutuviin jatkotoimenpiteisiin. Tuntemalla yleisön vältetään väärinymmärryksiä, tiedetään mitä varten visualisointi tehdään sekä mitä siitä tulee oppia. Näin ollen visualisoinnin jälkeen tehdyt toimenpiteet ovat laadukkaampia. Mielestäni yleisön tunteminen vaikuttaa myös kahteen muuhun yleiseen ohjeeseen (aineiston keskeisten piirteiden ja liian yksityiskohtainen kuvaaminen). Lisäksi tietämällä yleisön osaamisen ja tietämyksen sekä yleisön tietotarpeet, pystytään valitsemaan sopiva lähestymistapa visualisoinnin laajuudelle.
